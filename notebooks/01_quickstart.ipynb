# maritime-co2 quickstart: OPS-avoidable CO2 (liquid bulk)

Estimates the at-berth CO2 that Onshore Power Supply would avoid for tanker calls.

**Auxiliary engine only; boilers excluded** (they keep running for cargo heating).
Sample: 100 real Port of Itaqui liquid-bulk calls, 2024 (~10% of annual calls).

```
import pandas as pd
from maritime_co2 import auxiliary_emissions, estimate_ops_savings
```

## 1. Single call

```
r = auxiliary_emissions(power_kw=1000, load_factor=0.5, hours=24, fuel='MDO')
print(f'Auxiliary CO2 avoided by OPS: {r.co2_tonnes:.2f} t')
```

## 2. Real 2024 sample

```
calls = pd.read_csv('../data/sample/itaqui_liquid_bulk_2024_sample.csv')
res = estimate_ops_savings(calls)
res[['call_id','dwt','size_band','berth_hours','co2_avoided_tonnes']].head(10)
```

## 3. Totals and distribution

```
print(f"Calls: {len(res)} | Total OPS-avoidable CO2: {res['co2_avoided_tonnes'].sum():.1f} t")
res.groupby('size_band')['co2_avoided_tonnes'].sum().plot(kind='bar', title='OPS-avoidable CO2 by size band (t)')
```
